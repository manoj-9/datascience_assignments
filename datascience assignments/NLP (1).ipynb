```python
book = [x.strip() for x in book.x] #remove both the leading and the trailing characters
book = [x for x in book if x] #remove empty strings,because they are considered in python as false.
book[0:10]
```

```python
# joining the list into one text
text = " ".join(book)
text
```

```python
t='ki!#sq%'
to_remove=['!','#','%']
for each in to_remove:
  t=t.replace(each,'')
t
```

```python
#punctuations
no_punc_text=text.translate(str.maketrans('','',string.punctuation))#with argument(x,y,z)where x and y
#must be equal -length string amd character in 'x'
#are replaced by character in y and z
#is string (string.punctuation here)
no_punc_text

```

```python
import nltk
nltk.download('punkt')
```

```python
#tokenization
import nltk
from nltk.tokenize import word_tokenize
text_tokens = word_tokenize(no_punc_text)
print (text_tokens[0:5])
```

```python
list(no_punc_text)
```

```python
#remove stop words
import nltk
from nltk.corpus import stopwords
nltk.download('punkt')
nltk.download('stopwords')
my_stop_words =stopwords.words('english')
my_stop_words.append('the')
my_stop_words.append('Its')
no_stop_tokens = [word for word in text_tokens if not word in my_stop_words]
print(no_stop_tokens[0:40])
```

```python
#normalization the data
lower_words=[x.lower()for x in no_stop_tokens]
print(lower_words[0:25])
```

```python
#stemming
from nltk.stem import PorterStemmer
ps = PorterStemmer()
stemmed_tokens=[ps.stem(word)for word in lower_words]
print(stemmed_tokens[0:10])
```

```python
#nlp english langauge model of spacy library
nlp=spacy.load('en_core_web_sm')
```

```python
#lemmas being one of them,but mostly POS,which will follow later
doc=nlp(' '.join(no_stop_tokens))
print(doc[0:40])
```

```python
lemmas=[token.lemma_ for token in doc]
print(lemmas[0:40])
```

```python
#featire extraction
from sklearn.feature_extraction.text import CountVectorizer
vectorizer=CountVectorizer()
X=vectorizer.fit_transform(lemmas)
```
